### Comparison 

**Comaring** gives a measurement **how much our predictions missed by**.<br>
One of the most important problem in DL is choosing the error function.<br>
Given some predictions, we will calculate the error, saying whether **the predicted value is too high or too low**, compared to actual values. It does not say why it's too low or too high and the direction you missed, but will calculate the just the difference.

Within this chapter we'll apply MSE as the way of measureing the error forr regression problem (predicting the numerical value).

We will use **Gradient descent** as the method of tuning weights. It computes a number number for each weight.<br>
This number tells us how the weight should be **adjusted in order to reduce the error**.

### Mean squared error implementation

The error is the measure of how much our predictions have missed. Let's implement **MSE**.

```python
knob_weight = 0.5 # weight for our feature
inputs = 0.5 # the value of our feature
goal_pred = 0.8 # actual value or target 

pred = knob_weight * inputs
# calculate the error
error = (pred - goal_pred)**2
print(error)
```

Thus, MSE makes the error posses 2 properties - **being squared and being positive**.

#### Why do we square the error? 

The reason for this is making our error **positive**.<br>
For example, in the first iteration we've predicted the value **below target** by 2. In the second iteration you predicted **above taget** by 2 values.<br>
Squaring the error tells us **how much we missed by in either directions**|.

**Interesting note**

Dealing with errors > 1, we **amplify** errors, making them even bigger, but dealing with errors < 1, we **decreasing them**.<br>
But that's okay, because in practice we pay more attention to bigger errors. <br>
Applying MSE (squaring) we **prioritise bigger errors** and treat small errors as irrelevant. <br>
E.g. if the error is far less than 1, let's say 0.01, we say that the squared value (0.01 * 0.01 = 0.0001) becomes irrelevant.

#### Why do we need error to be positive? 
<br>
We are interested in calculating the average error across all data samples and making it down to 0.<br>

If we denote the error of batch `m` by `J` the average error will `sum(J)/m`.<br>
For example, we make predictions for 2 datapoint. <br>
Predicted value for the first one is equal to 1000, predicted value for the second is -1000.<br> 
In case we use just averaging, the **result will be missleading**: (1000-1000)/2 = 0, that means predictions are perfect.

### Simplest way of learning: 'Hot and cold' learning method

Given with specified **learning rate** `lr`, we can try to tune the weight up and down, so the error is reduced.<b>

We will do this many times and when the error is 0, we are done with learning.<br>
On each iteration we will try change weight **up and down** and check **which one (up or down) will reduce the error**.

```python
# define the function for neural network
def neural_network(inputs, weights):
    pred = inputs * weights
    return pred
```

```python
inputs = 0.5 # average number of toes for a single sample
goal_prediction = 0.8 # actual value 
weight = 0.5
lr = 0.001 # learning rate or change per one step
num_iter = 1011 # let's try this number of iterations

for i in range(num_iter):
    pred = neural_network(inputs, weight)
    error = (pred - goal_prediction)**2
    print('Error: ', str(error) + " Predictions: "+ str(pred))
    
    # now, try to change weight up and down
    pred_up = inputs * (weight+lr) # up weight
    pred_down = inputs * (weight-lr) # down weight
    # calculate 2 new errors
    error_up = (pred_up  - goal_prediction)**2
    error_down = (pred_down - goal_prediction)**2
    
    # if the error with higher weight is bigger than error with down weight
    # we turn the weight down and use smaller weight to make predictions
    if(error_down < error_up):
        weight = weight - lr
    # otherwise
    if(error_down > error_up):
        weight = weight + lr
        
```

As we can see from results above, **the error keeps decreasing** along with number of iterations. <br>
The last step is almost correctly predicts the actual value: `pred = 0.76` and `goal_prediction = 0.8`.

### Main problems with 'Hot and cold' technique

#### Problem 1: it's inefficient

We need to compute the error 3 times per iteration.<br>
The first error - with initial weights, then 2 errors more - with both up and down weights.<br>
Moreover, we need to do this for a **single weight** to be updated.

#### Problem 2: sometimes it's impossible to predict the exact target

Basically, given with the learning rate `lr`, neural network will eventually overshoot by some value less than `lr`, unless the optimal weight is exactly `num_iterations * lr` away from initial weight.<br>
The bigger problem is even though you know the right direction of the weight's change, you do not know the **correct amount of change - so the optimal learning rate**. Whether the error is big or not, the learning rate `lr` will be the same.

What if we find a way to compute **both direction and amount of weight**, without having to repeatedly making predictions?

### Gradient descent

Gradient descent is a method, that allows you to compute **both direction and amount of change**.

By definition (non-formal) the **gradient (or derivative)** is a slope of the function and represent the **change of dependent variable** with respect to independent (adjusting independent variable).<br>
Assume our error function `J` is the error function. Suppose it depends on the weight `w`.<br>
Then, if we take derivative of `J` with respect to `w` at specific point, this will represent the rate of change of `J` with respect to `w` at this point.

Note, if the gradient is **positive**, then the function shows **growth**.<br>
Otherwise, if it is **negative**, the function **decreases**. 

Hence, we can update our weight `w` in order to **minimize** the function `J`.

Formally, we can write the new weight as `w_new` = `w` - `lr` * `dw`, where `dw` - derivative of `J` with respect to `w`.

**How to compute derivative `dw`?**

Denote `a` to be predictions of neural network.<br>
Suppose, we do not use bias for now and compute predictions `a = activation(w * input)`, where `activation` can be any non-linear activation function (sigmoid, tangent, ReLU and etc.)<br>
Let's also denote `z = w * input` - lenear activation.

In order to find derivative `dw`, we need to perform the following steps:

 * `dz = a -  y`, where `a` - predictions and `y` - actual value.
 * `dw += inputs * dz`<br>
 Finally, we calculate error function `J` as taking an average `J/=m` and derivative `dw /=m` across all samples `m`.

In the current example we have `m=1`. <br>
After drivative calculation, we adjust our weight, using formula above `w_new = w - lr * dw`.<br>
Numpy non-vectorized code is shown below.

```python
weight = 0.5
goal_pred = 0.8
inputs = 0.5
num_iter = 20

for i in range(num_iter):
    pred = weight * inputs
    error = (pred - goal_pred)**2
    # for now, let's not use learning rate for simplicity
    # dz = (pred - goal_pred) = a-y
    # dw = dz * inputs
    dw = (pred - goal_pred)*inputs
    weight = weight - dw
    print('Error: '+str(error), 'Predictions: '+str(pred))
```

As you can see from above, our desired prediction value **improved significantly**.<br>
`dw` represents the amount and direction. Multiplying it by inputs provides us **scaling** ability, negative reversal and stopping.<br>
**Pure error** `(pred - goal_pred)` or `dz` shows us the raw direction and amount we missed with our predictions. If value is too big, we missed a lot. If value is *positive*, we predicted the way too high and vice versa. 

#### What are scaling, negative reversal, and stopping?


These attributes help to modify **pure error** to become the absolute value by which we adjust the weight.

**Stopping**

Stopping is an effect from multiplying pure error `dz` or `(pred - goal_pred)` by `inputs`.<br>
If `inputs` = 0, then there is nothing to learn. So, the `dw` will be 0 and `weiht` will stay the same.

**Negative reversal**

Let's split this effect into 2 parts:
 > a) If `inputs` > 0.<br>
 Then, if `(pred - goal_pred)` or `dz` > 0, we have `dw` > 0 and `weight` will be decreased.<br>
 This is because when `dw` is potitive, it means the error function is growing at this point. We want it to be minimised. So, we decrease our weight.<br>
 
 >b)  If `inputs` < 0. 
 Then, if `(pred - goal_pred)` or `dz` > 0, we have `dw` < 0 and `weight` will be increased (minus negative value).<br>
 So, in this case `weight` suddenly changes its direction and when we calculate predictions `pred = weight * inputs`, this will make predictions go down since we multiply big `weight` by negative `inputs`.

**Scaling**

Scaling just adresses the fact that we muliply **pure error** `dz` by `weights`.<br>
If the input is large, our weight change (`dw`) will be large as well. <br>
In order to control this problem, we will use `alpha` or `lr` - learning rate, ensuring the weight change has a reasonable value. 

Let's add learning rate `lr` to our grdient descent algorithm and call `delta` or `dz` - the difference between predictions and actual values. Now, `weight_delta` or `dw` will be the measure of how much we change our weight. We multiply it by `lr` to control the value of change.

```python
weight = 0.5
goal_pred = 0.8
inputs = 0.5
num_iter = 50000 # can be increased
lr = 0.01

for i in range(num_iter):
    pred = weight * inputs
    error = (pred - goal_pred)**2
    # dz = (pred - goal_pred) = a-y
    # dw = dz * inputs
    delta = (pred - goal_pred)
    weight_delta = delta*inputs
    
    weight -= dw * lr
    print('Error: '+str(error), 'Predictions: '+str(pred))
```
